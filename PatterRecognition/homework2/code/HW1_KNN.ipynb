```python
import numpy as np
import matplotlib.pyplot as plt

def pdf(x):
    y = np.zeros_like(x)
    y[x < -2] = 1 / (6 * np.abs(x[x < -2] + 4))
    y[(x >= -2) & (x < 3)] = 1 / (6 * np.abs(x[(x >= -2) & (x < 3)]))
    y[x > 3] = 1 / (6 * np.abs(x[x > 3] - 6))
    return y

x = np.linspace(-10, 10, 1000)
y = pdf(x)

plt.plot(x, y, label=r'$\hat{p}(x)$')
plt.xlabel('x')
plt.ylabel(r'$\hat{p}(x)$')
plt.title('Probability Density Function')
plt.legend()
plt.grid(True)
plt.savefig('knn.svg', format='svg')
plt.show()
```
