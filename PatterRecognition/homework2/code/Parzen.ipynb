```python
import numpy as np
import matplotlib.pyplot as plt

# 样本点
samples = [4.6019, 5.2564, 5.2200, 3.2886, 3.7942,
           3.2271, 4.9275, 3.2789, 5.7019, 3.9945,
           3.8936, 6.7906, 7.1624, 4.1807, 4.9630,
           6.9630, 4.4597, 6.7175, 5.8198, 5.0555,
           4.6469, 6.6931, 5.7111, 4.3672, 5.3927,
           4.1220, 5.1489, 6.5319, 5.5318, 4.2403,
           5.3480, 4.3022, 7.0193, 3.2063, 4.3405,
           5.7715, 4.1797, 5.0179, 5.6545, 6.2577,
           4.0729, 4.8301, 4.5283, 4.8858, 5.3695,
           4.3814, 5.8001, 5.4267, 4.5277, 5.2760]

def gaussian_window(x, h):
    return np.exp(-x**2 / (2*h**2)) / np.sqrt(2*np.pi*h**2)

def square_window(x, h):
    return (np.abs(x/h) <= 0.5)/h

def parzen_estimate(samples, x, window_function, h):
    n = len(samples)
    return sum([window_function((x-xi), h) for xi in samples]) / n

window_widths = [0.5, 1.0, 1.5, 2, 2.5]
x_points = np.linspace(2, 8, 1000)

plt.figure(figsize=(12, 6))
for h in window_widths:
    y_gaussian = [parzen_estimate(samples, x, gaussian_window, h) for x in x_points]
    plt.plot(x_points, y_gaussian, label=f'Gaussian h={h}')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Parzen Window Density Estimation - Gaussian Window')
plt.legend()
plt.grid(True)
plt.show()

# Plot for Square window
plt.figure(figsize=(12, 6))
for h in window_widths:
    y_square = [parzen_estimate(samples, x, square_window, h) for x in x_points]
    plt.plot(x_points, y_square, label=f'Square h={h}')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Parzen Window Density Estimation - Square Window')
plt.legend()
plt.grid(True)
plt.show()
```
