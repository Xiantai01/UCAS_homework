```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = 2 * np.pi  # 常数 w
T = 10.0  # 模拟的时间长度
dt = 0.01  # 时间步长
t = np.arange(0, T, dt)  # 时间数组

# 生成泊松过程 N(t)（用泊松分布的等待时间模拟）
lambda_poisson = 1.0  # 泊松过程的参数 lambda
num_events = np.random.poisson(lambda_poisson * T)  # 在 T 时间内发生的事件数量
event_times = np.sort(np.random.uniform(0, T, num_events))  # 事件发生的时间点

lambda_poisson1 = 1.0  # 泊松过程的参数 lambda
num_events1 = np.random.poisson(lambda_poisson * T)  # 在 T 时间内发生的事件数量
event_times1 = np.sort(np.random.uniform(0, T, num_events)) 
# 计算泊松过程 N(t) 在每个时间点的值
N_values = np.zeros_like(t, dtype=int)
for i, ti in enumerate(t):
    N_values[i] = np.sum(event_times <= ti)
    
N_values1 = np.zeros_like(t, dtype=int)
for i, ti in enumerate(t):
    N_values1[i] = np.sum(event_times1 <= ti)

# 计算随机过程 X(t)
X = np.cos(w * t + np.pi * N_values)
X1 = np.cos(w * t + np.pi * N_values1)

# 绘制样本轨道
plt.figure(figsize=(8, 5))
plt.plot(t, X, label=r'$X(t) = Acos(\omega t + \pi N(t))$', color='blue')
plt.plot(t, -X1, label=r'$X(t) = Acos(\omega t + \pi N(t))$', color='red')
plt.title('Sample Path of the Stochastic Process')
plt.xlabel('Time')
plt.ylabel('X(t)')
plt.legend(loc='upper left')
plt.grid(True)
plt.savefig('sample.svg',format='svg')
plt.show()
```
